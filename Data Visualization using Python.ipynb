```python
from sklearn.datasets import fetch_openml
boston = fetch_openml(name='boston', version=1, as_frame=True)
df = boston.frame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop rows with missing values for visualization purposes
df_clean = df.dropna()

# Set up plotting style
sns.set(style="whitegrid")

# Create visualizations
plt.figure(figsize=(16, 12))

# 1. Histogram of MEDV (Target Variable)
plt.subplot(2, 2, 1)
sns.histplot(df_clean['MEDV'], kde=True, color='skyblue')
plt.title('Distribution of Median Home Value (MEDV)')
plt.xlabel('Median Value ($1000s)')

# 2. Scatter plot: RM vs MEDV
plt.subplot(2, 2, 2)
sns.scatterplot(x='RM', y='MEDV', data=df_clean, hue='CHAS', palette='Set1')
plt.title('Number of Rooms vs Median Home Value')
plt.xlabel('Average Number of Rooms')
plt.ylabel('Median Value ($1000s)')

# 3. Heatmap of Feature Correlation
plt.subplot(2, 2, 3)
corr = df_clean.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')

# 4. Boxplot: LSTAT vs MEDV
plt.subplot(2, 2, 4)
sns.scatterplot(x='LSTAT', y='MEDV', data=df_clean, color='darkgreen')
plt.title('% Lower Status Population vs Median Home Value')
plt.xlabel('% Lower Status of Population')
plt.ylabel('Median Value ($1000s)')

plt.tight_layout()
plt.show()
```
